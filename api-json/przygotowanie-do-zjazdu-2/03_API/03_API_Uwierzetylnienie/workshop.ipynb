### Informacje o artyście, płytach i utworach

Przejdź do [theaudiodb.com/api_guide.php](https://theaudiodb.com/api_guide.php). Zapoznaj się z endpointami:

- `Return Artist details from artist name` (tu poznasz jego `idArtist`, czyli `TADB_Artist_ID`)
- `Return All Albums for an Artist using known TADB_Artist_ID` (tu poznasz `idAlbum`, czyli `TADB_Album_ID`)
- `Return All Tracks for Album from known TADB_Album_ID` (tu poznasz nazwy piosenek)

Użyj nazwy, aby dowiedzieć poznać `idArtist` za pomocą pierwszego endpointu.

Następnie użyj `idArtist`, aby poznać wszystkie `idAlbum` jego płyt.

Na końcu użyj `idAlbum`, aby poznać wszystkie piosenki jakie występują na danej płycie.

### Pobieranie tekstów piosenek

Do pobrania tekstów piosenek użyj [lyrics.ovh](https://lyrics.ovh/) - w szczególności zobacz [dokumentację jego jedynego endpointu](https://lyricsovh.docs.apiary.io/#reference/0/lyrics-of-a-song/search).

### Wyszukiwanie lokalizacji w tekście

Odebrany tekst piosenki możesz przekazać do API [geocode.xyz](https://geocode.xyz/api). Przykłady tam zamieszczone używają polecenia `curl`, dostępnego w systemach Linux i MacOS.

Przykład potrzebnego polecenia `curl` - poniżej znajdziesz to samo zapytanie wykonane za pomocą Pythona:

```bash
curl -X POST -d scantext="In the bestselling novel Crazy Rich Asians, author Kevin Kwan fills his pages with juicy drama involving Singaporeas elite families, and he almost devotes the same amount of space to the Southeast Asian nationas mouthwatering food." \
        -d geoit="json" \
        -d sentiment="analysis" \
        https://geocode.xyz
```

Python:

```python
text = """
In the bestselling novel Crazy Rich Asians, author Kevin Kwan 
fills his pages with juicy drama involving Singaporeas elite 
families, and he almost devotes the same amount of space to the 
Southeast Asian nationas mouthwatering food.

response = requests.post(

```
